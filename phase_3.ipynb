## Advance Operation with Business examples


```python
import numpy as np  
import matplotlib.pyplot as plt
#import pandas as pd
```

```python
# Data structure: [ resturant_id, 2021, 2022, 2023, 2024 ]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biriyani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 100000, 120000, 140000, 160000], # KFC
    [4, 90000, 100000, 110000, 120000], # McDonalds
    [5, 80000, 90000, 100000, 110000], # Burger King
    [6, 70000, 80000, 90000, 100000], # Pizza Hut
    [7, 60000, 70000, 80000, 90000], # Domino's
    [8, 50000, 60000, 70000, 80000], # Subway
    [9, 40000, 50000, 60000, 70000], # Taco Bell
    [10, 30000, 4000000, 50000, 60000] # Wendy's
    ])
print("==== Zomato Sales Data Analysis ====")
print("\n Sales Data Shape: ", sales_data.shape)
print('\n Sample Data for 1st 3 resturants: \n ', sales_data[:])
```

```python
# Total sales per year
total_sales = np.sum(sales_data, axis=0)
yearly_sales = np.sum(sales_data[:,1:], axis = 0)
print("\n Total Sales Per Year: \n ", total_sales)
print("\n Yearly Sales: \n ", yearly_sales)


# Minimum sales per resturant 
min_sales = np.min(sales_data[:,1:],axis=1)
# axis = 1 means row wise
# axis = 0 means column wise
print("\n Minimum Sales Per Resturant: \n ", min_sales)

# Maximum sales per year
max_sales =  np.max(sales_data[:,1:], axis = 0)
print("\n Maximum Sales Per Year: \n ", max_sales)

# Average Sales per resturant 
avg_sales = np.mean(sales_data[: ,1:], axis = 1 )
print("\n Average Sales Per Resturant: \n ", avg_sales)
```

```python
# Cumulative Sales per resturant
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print("\n Cumulative Sales Per Resturant: \n ", cumsum)

plt.figure(figsize=(10,5))
plt.plot(np.mean(cumsum, axis=0))
plt.title('Average Cumulative Sales Per Year across all Resturants')
plt.xlabel('Year')  
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
vector1 =np.array([1,2,3,4,5])
vector2 =np.array([6,7,8,9,10])
print('vector addition: ' , vector1 + vector2)
print('\nVector Multiplication : ', vector1 * vector2)
print('\nDot Product', np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print('\nAngle between vectors: ', angle)
```
